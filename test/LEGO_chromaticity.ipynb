```python
%matplotlib inline

import csv

import colour
import numpy as np
from colour.plotting import *

import matplotlib.colors as mc
```

```python
points = {}
hex_values = {}

with open('data/lego_colors.csv', mode='r') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    for row in csv_reader:
        try:
            name = row['name']
            hexvalue = "#"+row['rgb']
            rgb = mc.hex2color(hexvalue)
            print(name, rgb)
            points[name] = rgb
            hex_values[name] = hexvalue
        except ValueError:
            print("Invalid RGB: ", row['rgb'], 'for', row['name'])

incl_list = ["Bright Redish Violet", 
             "Bright Purple",
             "Light Purple"]



incl_list = ["Bright Redish Violet", "Vibrant Yellow"]




RGB_list = []
for key in points:
    if key in incl_list:
        print(key, hex_values[key])
        RGB_list.append(np.array(points[key]) / 255)

    if "Vibrant Yellow" in key:
        print(key, hex_values[key])
        RGB_list.append(np.array(points[key]) / 255)

        
plot_RGB_chromaticities_in_chromaticity_diagram_CIE1931(RGB_list, scatter_kwargs={"s": 8, "c":"RGB"})
```
